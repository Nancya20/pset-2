# Patrón Template Method (Método Plantilla)

## 1) ¿Qué problema resuelve?
Cuando tienes un algoritmo con pasos fijos (estructura estable), pero algunos pasos varían.

Ejemplo:
- Un proceso de "checkout" siempre:
  1) validar
  2) cobrar
  3) confirmar
  4) notificar
Pero “validar” o “notificar” puede variar por tipo de cliente/canal.

Si lo haces con copypaste:
- duplicas pasos,
- creas divergencias,
- y arreglar un bug requiere cambiar múltiples lugares.

---

## 2) Idea central
Una clase base define el "esqueleto" del algoritmo en un método:
- `template_method()`

y deja a subclases implementar los pasos variables mediante:
- métodos abstractos (`@abstractmethod`) o hooks (métodos opcionales).

---

## 3) Componentes
- **AbstractClass**: contiene el Template Method y pasos comunes.
- **ConcreteClass**: implementa pasos variables.

---

## 4) Cuándo usarlo (señales)
- Flujo casi idéntico en múltiples clases/variantes.
- Hay pasos invariantes + pasos variables.
- Quieres estandarizar el proceso evitando duplicación.

---

## 5) Beneficios
- Reduce duplicación.
- Asegura orden correcto del proceso.
- Facilita extender variaciones (subclases) sin romper el flujo.

---

## 6) Trade-offs
- Basado en herencia (puede ser rígido si hay muchas combinaciones).
- Para variación por composición, Strategy a veces es mejor.

---

## 7) Diferencia con Strategy
- **Template Method**: mismo algoritmo base, varían pasos en subclases.
- **Strategy**: intercambias un algoritmo completo por composición.
